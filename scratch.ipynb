```python
import numpy as np
import matplotlib.pyplot as plt
from time import sleep
```

```python
%matplotlib notebook
fig, ax = plt.subplots()
for n in [1, 2, 3] :
    x = np.arange(0,n,0.01)
    y = np.sin(x)
    if ax.lines :
        ax.lines[0].set_data(x, y)
        delta_x = 0.05*(max(x)-min(x))
        delta_y = 0.05*(max(y)-min(y))
        ax.set_xlim(min(x)-delta_x, max(x)+delta_x)
        ax.set_ylim(min(y)-delta_y, max(y)+delta_y)
    else :
        plt.plot(x, y)
    fig.canvas.draw()
    sleep(1)
```
